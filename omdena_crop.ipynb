```python
#retaining temp, weather code, daylight,precipitation_sum, region, longitude, latitude
climate_new = climate[['date','weather_code','temperature_2m_mean','precipitation_sum','apparent_temperature_mean','region', 'latitude', 'longitude']]
climate_new = climate_new.dropna()
```

```python
#aggregate climate data to months

# Custom mode function that returns the first mode if multiple modes exist
def get_mode(series):
    return series.mode()[0] if not series.mode().empty else None


# Convert the 'date' column to datetime
climate_new['date'] = pd.to_datetime(climate_new['date'])

# Extract year and month
climate_new['year'] = climate_new['date'].dt.year
climate_new['month'] = climate_new['date'].dt.month

# Group by the necessary columns
climate_agg = climate_new.groupby(['year', 'month', 'region', 'latitude', 'longitude']).agg({
    #'weather_code': get_mode,
    'weather_code': 'mean',
    'temperature_2m_mean': 'mean',
    'precipitation_sum': 'mean',
    #'apparent_temperature_mean': 'mean',
}).reset_index()
```

```python
#drop 1940, east climate data starts at 1950
climate_agg = climate_agg.loc[~climate_agg['year'].between(1940, 1949)]
climate_agg = climate_agg.iloc[:-4]
climate_agg.head()
```

```python
#aggregate climate data to months

# Custom mode function that returns the first mode if multiple modes exist
def get_mode(series):
    return series.mode()[0] if not series.mode().empty else None


# Convert the 'date' column to datetime
east_climate['date'] = pd.to_datetime(east_climate['Week Start Date'])

# Extract year and month
east_climate['year'] = east_climate['date'].dt.year
east_climate['month'] = east_climate['date'].dt.month

# Group by the necessary columns
east_climate_agg = east_climate.groupby(['year', 'month', 'region', 'latitude', 'longitude']).agg({
    #'weather_code': get_mode,
    'weather_code': 'mean',
    'temperature_2m_mean': 'mean',
    'precipitation_sum': 'mean',
    #'apparent_temperature_mean': 'mean',
}).reset_index()

east_climate_agg
```

```python
climate_new_agg['weather_code'].unique() #---> can be used to determine if dry or wet season
```

```python
# Function to classify season based on weather code
def classify_season(code):
    if code >= 50:  # Assuming codes 50 and above are "wet"
        return 'wet'
    else:
        return 'dry'

# Apply the function to the 'weather_code' column and create a new 'season' column
climate_new_agg['season'] = climate_new_agg['weather_code'].apply(classify_season)

# Display the updated DataFrame
climate_new_agg.tail()
```
